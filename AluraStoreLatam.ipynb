### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
total_ganancias= tienda["Precio"].sum()
total_ganancias2= tienda2["Precio"].sum()
total_ganancias3= tienda3["Precio"].sum()
total_ganancias4= tienda4["Precio"].sum()
```

# 2. Ventas por categoría

```python
ventas_por_categoria=tienda.groupby("Categoría del Producto")['Producto'].count()
ventas_por_categoria2=tienda2.groupby("Categoría del Producto")['Producto'].count()
ventas_por_categoria3=tienda3.groupby("Categoría del Producto")['Producto'].count()
ventas_por_categoria4=tienda4.groupby("Categoría del Producto")['Producto'].count()
```

```python
ventas_por_categoria
```

# 3. Calificación promedio de la tienda


```python
tienda.sample(10)
```

```python
calificacion_promedio=tienda['Calificación'].mean()
calificacion_promedio2=tienda2['Calificación'].mean()
calificacion_promedio3=tienda3['Calificación'].mean()
calificacion_promedio4=tienda4['Calificación'].mean()

```

# 4. Productos más y menos vendidos

```python
ventas_totales_por_producto = tienda["Producto"].value_counts()
ventas_totales_por_producto2 = tienda2["Producto"].value_counts()
ventas_totales_por_producto3 = tienda3["Producto"].value_counts()
ventas_totales_por_producto4 = tienda4["Producto"].value_counts()


```

```python

producto_mas_vendido = ventas_totales_por_producto.idxmax()
producto_menos_vendido = ventas_totales_por_producto.idxmin()

producto_mas_vendido2 = ventas_totales_por_producto2.idxmax()
producto_menos_vendido2 = ventas_totales_por_producto2.idxmin()

producto_mas_vendido3 = ventas_totales_por_producto3.idxmax()
producto_menos_vendido3 = ventas_totales_por_producto3.idxmin()

producto_mas_vendido4 = ventas_totales_por_producto4.idxmax()
producto_menos_vendido4 = ventas_totales_por_producto4.idxmin()

```

# 5. Envío promedio por tienda

```python
costo_envio_promedio=tienda['Costo de envío'].mean()
costo_envio_promedio2=tienda2['Costo de envío'].mean()
costo_envio_promedio3=tienda3['Costo de envío'].mean()
costo_envio_promedio4=tienda4['Costo de envío'].mean()
```

## *6. Representación de los datos**

```python
df_resumen = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ganancias Totales': [total_ganancias, total_ganancias2, total_ganancias3, total_ganancias4],
    'Calificación Promedio': [calificacion_promedio, calificacion_promedio2, calificacion_promedio3, calificacion_promedio4],
    'Costo Envío Promedio': [costo_envio_promedio, costo_envio_promedio2, costo_envio_promedio3, costo_envio_promedio4]
}).set_index('Tienda')

```

```python
df_resumen['Ganancias Totales'].plot(kind='bar', color='skyblue', edgecolor='black', figsize=(7,5))
plt.title('Ingresos Totales por Tienda')
plt.ylabel('Monto en $')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
df_resumen['Calificación Promedio'].plot(kind='line', marker='o', color='orange', figsize=(7,5))
plt.title('Satisfacción de Clientes por Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(0, 5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
ventas_por_categoria.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6,6))
plt.title('Distribución de Categorías de Productos - Tienda 1')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
df_resumen.plot(kind='scatter', x='Ganancias Totales', y='Costo Envío Promedio', color='purple', s=120, figsize=(7,5))
plt.title('Relación entre Ingresos y Costo Promedio de Envío')
plt.xlabel('Ganancias Totales ($)')
plt.ylabel('Costo Promedio de Envío ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
